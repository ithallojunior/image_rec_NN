## Example of an artificial neural network (ann) in classifying an image of a raccoon.
#### Author: Ithallo J. A. Guimar√£es    Aug/2016

```python
from sklearn.neural_network import MLPClassifier
from scipy import misc
import numpy as np
import random as rd
import matplotlib.pyplot as plt
import pickle
```

```python
#loading images, removing one of the dimensions of the colors (making it black and white)
#Raccon images

y1 = misc.imread('./set/r2.png')
y2 = misc.imread('./set/r6.png')
y3 = misc.imread('./set/r3.png')
y4 = misc.imread('./set/r4.png')
y5 = misc.imread('./set/r5.png')
y6 = misc.imread('./set/r1.png')
y7 = misc.imread('./set/r7.png')
#other images
n0 = misc.imread('./set/chimp.png')
n1 = misc.imread('./set/not1.png')
n2 = misc.imread('./set/dog.png')
n3 = misc.imread('./set/not2.png')
n4 = misc.imread('./set/bullterrier.png')

raw_inputs = [np.array(y1[:,:,0]), np.array(y2[:,:,0]), np.array(y3[:,:,0]),
              np.array(y4[:,:,0]), np.array(y5[:,:,0]), np.array(y6[:,:,0]),
              np.array(y7[:,:,0]), np.array(n0[:,:,0]), np.array(n1[:,:,0]),
              np.array(n2[:,:,0]),np.array(n3[:,:,0]), np.array(n4[:,:,0])]
for i in raw_inputs:
    print i.shape
```

```python
%matplotlib inline
# plotting images
n = 1
if len(raw_inputs) <= 9:
    for i in xrange(1,10):
        plt.subplot(int("%d%d" %(33,n )))
        plt.imshow(raw_inputs[i-1], cmap=plt.cm.gray)
        plt.axis('off')
        n=n+1
    plt.show()
else:
    for i in xrange(0,9):
        plt.subplot(int("%d%d" %(33,n )))
        plt.imshow(raw_inputs[i], cmap=plt.cm.gray)
        plt.axis('off')
        n=n+1
    plt.show()
    n=1
    for i in raw_inputs[9:]:
        plt.subplot(int("%d%d" %(33,n )))
        plt.imshow(i, cmap=plt.cm.gray)
        plt.axis('off')
        n=n+1
    plt.show()
```

```python
#Resizing the images, as the inputs onf the ANN must have the same size
def new_img(photo, r=100, c=100):
    new = []
    otp = []
    rows = np.linspace(0,photo.shape[0],r).astype(int)
    columns = np.linspace(0,photo.shape[1], c).astype(int)
    for  i in rows-1:
        for j in columns-1:
            new.append(photo[i,j])
        otp.append([new]) 
        new = []
    return np.hstack(otp) 

#reducing images to only 100x100 pixels
X =[]
for i in raw_inputs:
    X.append(new_img(i))
X= np.vstack(X)
print X.shape

 
#Creating the training set, only with positive reinforcements
X_train = X[[0,1, 2 , 3, 4],:]
print X_train.shape
y_train = np.ones((X_train.shape[0]))
#y_train[-1] = 0
#y_train[-2] = 0
#y_train[-3] = 0
print y_train.shape

X_test = X[[5, 6, 7 , 8,9,10,11],:]
y_test = np.zeros((X_test.shape[0])) 
y_test[0] = 1
y_test[1] = 1
print X_test.shape, y_test.shape
```

```python
#The ANN itself, yes, it has too many layers
clf = MLPClassifier(activation='logistic', tol =1e-12, algorithm='l-bfgs', alpha=0.01, 
                    hidden_layer_sizes=(500,300), random_state=None, verbose=False, max_iter=8000, shuffle= False)
clf.fit(X_train, y_train)
```

```python
print"Accuracy on train set:", clf.score(X_train, y_train) 
print"Accuracy on test set:", clf.score(X_test, y_test) 
print "Total iterations", clf.n_iter_
```

```python
n=5
plt.imshow(raw_inputs[n], cmap=plt.cm.gray)
if clf.predict(X[n].reshape(1,-1))== 1: plt.title("it's a raccoon")
else: plt.title(" It's not a raccoon")
plt.show()
```
